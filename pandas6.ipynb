```python
import re
flight_5 = flight.set_index(['Year']).sort_index().loc[2017:2019].loc[:, ['Whole year','Airport coordinates']]

# 坐标 提取
longitude = []
latitude = []
for i in range(0, len(flight_5['Airport coordinates'])):
    string = re.findall(r"\d+\.?\d*", str(flight_5['Airport coordinates'].iloc[i]))
    
    if string==[]:
        longitude.append(np.nan)
        latitude.append(np.nan)
    else:
        longitude.append(float(string[0]))
        latitude.append( float(string[1]))

flight_5['Longitude'] = longitude
flight_5['Latitude'] = latitude
flight_5 = flight_5.dropna()   
```

```python
cuts_long = pd.cut(flight_5['Longitude'],bins=[0, flight_5['Longitude'].median(), 180])
flight_5['cuts_long'] = cuts_long
cuts_la = pd.cut(flight_5['Latitude'],bins=[0, flight_5['Latitude'].median(), 90])
flight_5['cuts_la'] = cuts_la
display(flight_5.groupby(['cuts_long', 'cuts_la']).count())
groupby_5 = flight_5.groupby(['cuts_long', 'cuts_la'])
print( '2017 年 - 2019 年货运总量最大区域的 经度范围：%s，维度范围：%s' \
      % (groupby_5['Whole year'].sum().idxmax()[0], groupby_5['Whole year'].sum().idxmax()[1]) )
```

```python
三、新冠肺炎在美国的传播
```

```python
df_death = pd.read_csv(r'C:\Users\de\Desktop\数据集\美国死亡数.csv')
df_diagnosis  = pd.read_csv(r'C:\Users\de\Desktop\数据集\美国确证数.csv')
```

```python
#(a)调用corr计算小镇人口与最后一天【死亡数】的相关系数
df_death[['Population','2020/4/26']].corr()
```

```python
#(2)截止到4月1日，统计每个州零【感染】县的比例（以下计算4月1日数据）
df_2 = df_diagnosis[['Admin2','Province_State','2020/4/1']]
(df_2[df_2['2020/4/1']==0].groupby('Province_State')['2020/4/1'].count()/df_2.groupby('Province_State')['2020/4/1'].count()).fillna(0)
```

```python
#(3) 请找出最早出确证病例的三个县。
df3 = df_diagnosis.copy()
columnsdate = df3.columns[11:];top3 = 3;toptown=[];starttime=[]
#print(columnsdate)
for date in columnsdate:
    town = df3[df3[date]>0]#选出top
    ts = town.shape[0]
    if ts>0:
        top3 -= ts
        starttime += [date for i in range(ts)]
        toptown +=list(town['Admin2'])
        df3   = df3.drop(index=town.index)
        if top3<=0:
            break
print('最早发生：',toptown)
print('首现日期：',starttime)
```

```python

#(4) 按州统计单日死亡增加数，并给出哪个州在哪一天确诊数增加最大（这里指的是在所有州和所有天两个指标一起算，不是分别算）。
df4 = pd.merge(df_death,df_diagnosis,on=df_death.columns[0:10].to_list(),suffixes=('_death','_diag'))
df4.iloc[:,0:15].head()
```

```python
df4_death = df_death.copy()
df4_diagnosis  = df_diagnosis.copy()
columnsdate = df4_death.columns[12:].to_list()
#print(columnsdate)
datelen = len(columnsdate)
#print(datelen)
for i in range(datelen-1,0,-1):
    ci=columnsdate[i];  ci_=columnsdate[i-1]
    df4_death[ci] = df4_death.apply(lambda x : x[ci]-x[ci_],axis=1)
    df4_diagnosis[ci]  = df4_diagnosis.apply(lambda x : x[ci]-x[ci_],axis=1)
```

```python
df4_death.groupby('Province_State')[columnsdate].sum()
```

```python
df4_diagnosis.groupby('Province_State')[columnsdate].sum().idxmax(axis=1)
```

```python
#(5)现需对每个州编制确证与死亡表，第一列为时间，并且起始时间为该州开始出现死亡比例的那一天，第二列和第三列分别为确证数和死亡数，每个州需要保存为一个单独的csv文件，文件名为“州名.csv”。
df5_death = df_death.copy()
df5_diagnosis  = df_diagnosis.copy()
death = df5_death.groupby('Province_State').sum().iloc[:,6:].T
diag  =  df5_diagnosis.groupby('Province_State').sum().iloc[:,5:].T
States = df5_death.groupby('Province_State')['Admin2'].sum().index.to_list()
for state in States:
    aaa = pd.DataFrame({'确证数':diag[state],'死亡数':death[state]})
    for index,row in aaa.iterrows():
        if row['死亡数']>0:
            break
        else:
            aaa.drop(index,inplace=True)
    aaa.to_csv(state+'.csv')
```

```python
#(6)现需对4月1日至4月10日编制新增确证数与新增死亡数表，第一列为州名，第二列和第三列分别为新增确证数和新增死亡数，分别保存为十个单独的 csv 文件，文件名为“日期.csv”。
April = columnsdate[-26:-16]
#利用（4）结论按州统计单日死亡、确证增加数
death6 = df4_death.groupby('Province_State')[April].sum()
#给出哪个州在哪一天确诊数增加最大
diag6   = df4_diagnosis.groupby('Province_State')[April].sum()
for day in April:
    aaa = pd.DataFrame({'确证数':diag6[day],'死亡数':death6[day]})
    #print(day,aaa.head())
    aaa.to_csv(day.replace('/', '_')+'.csv')
```
